```python
houses_sp.info()
```

```python
#remoção de colunas com dados constantes
houses_sp=houses_sp.drop(columns=['city'])
houses_sp.head()
```

```python
#Lidando com dados nulos 
houses_sp['rooms'].median()
```

```python
houses_sp['rooms']=houses_sp['rooms'].fillna(houses_sp['rooms'].median())
```

```python
houses_sp.info()
```

```python
houses_sp['floor'].value_counts()
```

```python
#substituição de informação - por 0
houses_sp.loc[houses_sp['floor'] == '-', 'floor'] = '0'
houses_sp
```

# Outliers

Dados que distoam do padrão e consequentemente geram uma distorção nas analises

```python
houses = pd.read_csv("houses_to_rent.csv")
houses.head()
```

```python
houses.describe()
```

```python
houses['bathroom'].value_counts()
```

```python
#Distancia inter quartil - IQR
q1= houses['bathroom'].quantile(0.25)
q3= houses['bathroom'].quantile(0.75)

IQR= q3-q1
f"IQR: {IQR}"
```

```python
#filtrando os outliers
houses_outliers = houses[(houses['bathroom']<q1-(IQR*1.5)) | (houses['bathroom']>q3+(IQR*1.5))]
houses_outliers
```

```python
#Criando a base sem outliers
houses_inliers =  houses[(houses['bathroom']>=q1-(IQR*1.5)) & (houses['bathroom']<=q3+(IQR*1.5))]
houses_inliers
```

```python
houses_inliers.describe()
```

# Visualização de dados

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris= sns.load_dataset('iris')
iris.head()
```

```python
#criando histograma da distribuição de uma variável

col='sepal_length'
sns.histplot(data=iris, x='sepal_length', kde=True).set_title(f'Distribuição da variável {col}')
```

```python
for col in iris.drop(columns='species'):
    sns.histplot(data=iris, x=col, kde=True).set_title(f'Distribuição da variável {col}')
    plt.show()
```

```python
for col in iris.drop(columns='species'):
    sns.histplot(data=iris, x=col, kde=True, hue=iris['species']).set_title(f'Distribuição da variável {col}')
    plt.show()
```

```python
for col in iris.drop(columns='species'):
    sns.barplot(data= iris, x=iris['species'], hue=iris['species'], y= col, ci= 95)
    plt.show()
```

# Visualização de séries temporais 


```python
candy=pd.read_csv('candy_production.csv')
candy.head(10)
```

```python
candy['observation_date']=pd.to_datetime(candy['observation_date'])
candy.info()
```

```python
candy.plot(x='observation_date',
           y= 'industrial_production',
           title='Série Temporal')
```

```python
candy_filtered= candy[candy['observation_date']>='2010-01-01']
candy_filtered.head()
```

```python
candy_filtered.plot(x='observation_date',
                    y= 'industrial_production',
                    title='Série Temporal',
                    figsize=(12,6))
```

```python
ax=candy_filtered.plot(x='observation_date',
                       y= 'industrial_production',
                       title='Série Temporal', figsize=(12,6))
xcoords=['2011-01-01',
        '2012-01-01',
        '2013-01-01',
        '2014-01-01',
        '2015-01-01',
        '2016-01-01',
        '2017-01-01']

for xc in xcoords:
    plt.axvline(x=xc,
                 color='black',
                 linestyle='--')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

candy_filtered.set_index('observation_date', inplace=True)

analysis = candy_filtered[['industrial_production']].copy()

#analise de decomposição da série temporal
decompose_result = seasonal_decompose(analysis, model='multiplicative')

trend = decompose_result.trend
seasonal = decompose_result.seasonal
residual = decompose_result.resid

decompose_result.plot()
```
